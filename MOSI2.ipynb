```python
import numpy as np
import pandas as pd
import math

def LCG(a, m, c, z0):
  while True:
    z0 = (a * z0 + c) % m
    yield z0 / m

def appendDictToDF(df,dictToAppend):
  df1 = pd.concat([df, pd.DataFrame.from_records([dictToAppend])])
  return df1

def main():
  list_of_data = [10, 13, 12, 10, 10, 12, 10, 10, 12, 11, 11, 13, 10, 13, 10, 13, 10, 11, 12, 11, 10, 13, 11, 12, 13,
                  10, 12, 10, 11, 13, 12, 13, 11, 12, 10, 10, 11, 12, 13, 10, 12, 10, 11, 13, 12, 11, 11, 13, 13, 10]
  df = pd.Series(list_of_data).value_counts().reset_index().sort_values('index').reset_index(drop=True)
  df.columns = ['Lama Pelayanan', 'Frekuensi']
  df['ΣXi'] = df['Lama Pelayanan'] * df['Frekuensi']
  df['ΣXi'] =  df['ΣXi'].sum()
  df['μ'] = df['ΣXi'] / 50
  df['σ'] = np.std(list_of_data)

  a = 10; m = 121; c = 194; z0 = 10121192
  generator = LCG(a, m, c, z0)
  list = []
  df1 = pd.DataFrame(columns=['i','Zi-1','Zi','Ui'])
  z_prev = z0
  u_prev = next(generator)
  for i in range(50):
    z_curr = (a * z_prev + c) % m
    ui = next(generator)
    list.append(z_prev)
    df1 = appendDictToDF(df1,{ 'i' : i+1, 'Zi-1': z_prev, 'Zi': z_curr, 'Ui+1': ui , 'Ui': u_prev})
    u_prev = ui; z_prev = z_curr

  for j in range(50):
    R = np.sqrt(-2 * np.log(df1['Ui']))
    Theta = 2 * np.pi * df1['Ui+1']
    X = R * np.cos(Theta)
    result = df['μ'] + df['σ'] * X
    df2 = pd.DataFrame(data=result).reset_index().drop(columns='index').dropna()
    df2.columns = ['Hasil']
  print(df2)

if __name__ == "__main__":
  main()
```
