```python
urls = []

#Validation Picture Links
urls.append('https://img2.pngdownload.id/20180218/orq/kisspng-table-wood-furniture-table-5a8921f9d9d568.9004836715189365698923.jpg')
urls.append('https://img2.pngdownload.id/20180523/cav/kisspng-table-garden-furniture-chair-vintage-bamboo-5b05fd20c4f117.6783158615271191368067.jpg')
urls.append('https://img2.pngdownload.id/20190506/pzh/kisspng-portable-network-graphics-clip-art-transparency-im-table-transparent-png-file-png-image-transparent-5ccff128e08cc5.9757280115571315609198.jpg')
urls.append('https://img2.pngdownload.id/20171217/dc5/wood-door-png-5a3685adce8842.320235691513522605846.jpg')
urls.append('https://img2.pngdownload.id/20171220/jhe/wood-door-png-5a3a57ce4e7056.88558196151377300632134560.jpg')
urls.append('https://img.iproperty.com.my/angel-legacy/1110x624-crop/static/2020/12/1.-Desain-Rumah-Minimalis-dengan-Atap-Pelana.jpg')

imgs = []
plt.figure(figsize=(15, 6))

#Plot Validation Image
for u in range(len(urls)):
  r = requests.get(urls[u], stream=True)
  img = np.array(Image.open(r.raw))
  plt.subplot(1,len(urls), u+1)
  plt.imshow(img)
  img = cv2.resize(img, (224, 224))
  imgs.append(img)
  plt.imshow(img)

imgs=np.array(imgs)
print(imgs.shape)
```

```python
yh = model.predict(imgs)

#Image Predictions
for i in range(len(urls)):
  if np.argmax(yh[i]) > 2 and np.argmax(yh[i]) < 619:
    print('Meja')
  elif np.argmax(yh[i]) > 630 and np.argmax (yh[i]) < 750:
    print('Pintu')
  else:
    print("Bukan pintu atau meja")
```
